```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from scipy.optimize import curve_fit

import sys
import os
sys.path.insert(0, os.path.abspath('../model'))

import Pacejka52 as pa52
import fiala as fi
import Brush as br
import TMeasy as tm
import Dugoff as du
import Pacejka52_glSA as pa52gl
import Pacejka52_Sv_Sh_0 as pa52_0

i=10000
Sx = np.linspace(-1,1,i)
Sy = np.linspace(0,0.3,i)

Fz0 = 3114
IA = 0

#Fz Bereich + Anzahl der Schritte
iter = 10
Fz = np.linspace(1500,5000,iter) 
```

```python
#lateral

#MFsimpel
Fy_params_init = [-9,1.6,1.5,-24]

fig,ax = plt.subplots(4, figsize = [10,20])
plt.figure(5)
#D Parameter
D_pa52_0 = np.zeros(len(Fz))
D_pa5 = np.zeros(len(Fz))
for i in range(0,len(Fz)):
   
    def func_Fy(Sy,B,C,D,E):
        return pa52_0.Pacejka5_lat_pa(Sy,Fz[i],B,C,D,E)

    y_data = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[0]
    D_pa52_0[i] = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[1] / Fz[i]
    
    Fy_params, _ = curve_fit(func_Fy, Sy, y_data, Fy_params_init)
    D_pa5[i] = abs(Fy_params[2])

    
    Fy_fit = func_Fy(Sy, *Fy_params)
    plt.plot(Sy,y_data)
    plt.plot(Sy,Fy_fit, 'r--' ) 
    plt.xlim(0,0.3)
    plt.xlabel('Schrglaufwinkel [rad]')
    plt.ylabel('Fy [N]')


#Plotting:
ax[0].plot(Fz,D_pa52_0, label = 'Pacejka52')
ax[0].plot(Fz,D_pa5, 'r--', label = 'Pacejka5')
ax[0].legend()
ax[0].set_xlabel('Fz [N]')
ax[0].set_ylabel('D Parameter')
ax[0].set_title('D Parameter')
ax[0].set_xlim(1500,5000)
   
#B Parameter
B_pa52_0 = np.zeros(len(Fz))
B_pa5 = np.zeros(len(Fz))
for i in range(0,len(Fz)):
   
    def func_Fy(Sy,B,C,D,E):
        return pa52_0.Pacejka5_lat_pa(Sy,Fz[i],B,C,D,E)

    y_data = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[0]
    B_pa52_0[i] = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[2] 
    
    Fy_params, _ = curve_fit(func_Fy, Sy, y_data, Fy_params_init)
    B_pa5[i] = Fy_params[0]


#Plotting:
ax[1].plot(Fz,B_pa52_0, label = 'Pacejka52')
ax[1].plot(Fz,B_pa5, 'r--', label = 'Pacejka5')
ax[1].legend()
ax[1].set_xlabel('Fz [N]')
ax[1].set_ylabel('B Parameter')
ax[1].set_title('B Parameter')
ax[1].set_xlim(1500,5000)

#C Parameter
C_pa52_0 = np.zeros(len(Fz))
C_pa5 = np.zeros(len(Fz))
for i in range(0,len(Fz)):
   
    def func_Fy(Sy,B,C,D,E):
        return pa52_0.Pacejka5_lat_pa(Sy,Fz[i],B,C,D,E)

    y_data = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[0]
    C_pa52_0[i] = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[3] #/ Fz[i]
    
    Fy_params, _ = curve_fit(func_Fy, Sy, y_data, Fy_params_init)
    C_pa5[i] = Fy_params[1]


#Plotting:
ax[2].plot(Fz,C_pa52_0, label = 'Pacejka52')
ax[2].plot(Fz,C_pa5, 'r--', label = 'Pacejka5')
ax[2].legend()
ax[2].set_xlabel('Fz [N]')
ax[2].set_ylabel('C Parameter')
ax[2].set_title('C Parameter')
ax[2].set_xlim(1500,5000)

#E Parameter
E_pa52_0 = np.zeros(len(Fz))
E_pa5 = np.zeros(len(Fz))
for i in range(0,len(Fz)):
   
    def func_Fy(Sy,B,C,D,E):
        return pa52_0.Pacejka5_lat_pa(Sy,Fz[i],B,C,D,E)

    y_data = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[0]
    E_pa52_0[i] = pa52_0.Pacejka52_lat(Fz[i],Sy,IA,Fz0)[4] 
    
    Fy_params, _ = curve_fit(func_Fy, Sy, y_data, Fy_params_init)
    E_pa5[i] = Fy_params[3]


#Plotting:
ax[3].plot(Fz,E_pa52_0, label = 'Pacejka52')
ax[3].plot(Fz,E_pa5, 'r--', label = 'Pacejka5')
ax[3].legend()
ax[3].set_xlabel('Fz [N]')
ax[3].set_ylabel('E Parameter')
ax[3].set_title('E Parameter')
ax[3].set_xlim(1500,5000)
```
